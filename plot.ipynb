```python
import pandas as pd

# Load data
df = pd.read_csv("output.txt")
df.rename(columns=lambda x: x.strip(), inplace=True)
```

```python
import matplotlib.pyplot as plt

i = 0
for initial_condition in df["initial_condition"].unique():
    ax = plt.figure(figsize=(10, 10)).add_subplot(projection="3d")
    dfi = df[df["initial_condition"] == initial_condition]

    xyzs = dfi[["x", "y", "z"]].to_numpy()
    ax.plot(*xyzs.T, lw=0.5)
    i += 1

    # Add legend
    x = dfi["x"].iloc[0]
    y = dfi["y"].iloc[0]
    z = dfi["z"].iloc[0]

    ax.set_title(f"Initial State: $x_0={x}$, $y_0={y}$, $z_0={z}$")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")

    plt.savefig(f"output_{i}.png")
```
